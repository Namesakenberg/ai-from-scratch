```python
# generating 2 dimensional dataset

import numpy as np
X = np.random.randn(100,2)
# make a ground truth for each point
# consider a line 
true_weights = np.array([[1],[-2]])
true_bias = 0.35
z = X@true_weights+true_bias
y = (z>0).astype(int)
```

```python
count_one,count_zero = 0,0
for i in y :
  if i[0] ==1 : 
    count_one+=1
  else:
    count_zero+=1

print(count_zero,count_one)
```

```python
class MyLogisticRegression:
  def __init__(self,learning_rate=0.1,epochs=100) -> None:
    self.lr = learning_rate
    self.epochs = epochs

  def sigmoid(self,z):  
    return 1/(1+np.exp(-z))

  def fit(self,X,y):
    self.n_samples = X.shape[0]
    self.n_features = X.shape[1]

    self.weights = np.random.randn(self.n_features)
    self.bias = 0

    for epoch in range(self.epochs):
      

      z = X@self.weights+self.bias
      y_hat = self.sigmoid(z)

      error = y_hat - y.flatten() 

      dw = (1/self.n_samples)* (X.T @ error)
      db = (1/self.n_samples)* np.sum(error)

      self.weights -= self.lr*dw
      self.bias -= self.lr*db

    return self.weights,self.bias

  def predict(self,X):
    # get the 
    z = X@self.weights+self.bias
    probs = self.sigmoid(z)
    predicted_labels= (z>0.5).astype(int)
    return predicted_labels
```

```python
# generate a test point and its actual class

import numpy as np
X_test = np.random.randn(50,2)
# make a ground truth for each point
# consider a line 
true_weights = np.array([[1],[-2]])
true_bias = 0.35
z = X_test@true_weights+true_bias
y_true = (z>0.5).astype(int)
```

```python
logreg = MyLogisticRegression()
logreg.fit(X,y)
y_pred = logreg.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_true,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X,y)
y_pred = lr.predict(X_test) 
print(accuracy_score(y_true,y_pred))
```
