```python
import numpy as np
```

```python
# coding batch gradient descent

 # considerinf y =  7x1-12x2+4x3-6

X = np.random.randn(100,3)
original_weights = np.array([[7],[-12],[4]])
original_bias = -6
noise = np.random.randn(100,1) * 0.5

y = X @ original_weights + original_bias + noise

ones = np.ones((100,1)) # bias = 1 initially
X_new = np.hstack((ones,X)) # training data

class MyBatchgradientDescent:
    def __init__(self,learning_rate = 0.01 ,epochs=1000):
        self.epochs = epochs
        self.lr = learning_rate
        self. theta = None


    def fit(self,X,y):
        n_samples = X.shape[0]
        ones = np.ones((n_samples,1))
        X_new = np.hstack((ones,X))
        n_features = X_new.shape[1]
        self.theta = np.random.randn(n_features,1)        # randomly initialize weights

        for i in range(self.epochs):
            y_hat = X_new@self.theta                      # make prediction
            error = y_hat-y                               # get the error
            gradient = (2/n_samples)*X_new.T @ error      # partial differentiation of loss wrt weights
            self.theta = self.theta - self.lr*gradient    # update rule

    def predict(self,X):
        n_samples = X.shape[0]
        ones = np.ones((n_samples, 1))
        X_new = np.hstack((ones, X))

        return X_new @ self.theta

    def get_weights(self):
        return self.theta
```

```python
bgd = MyBatchgradientDescent()
bgd.fit(X,y)
y_pred = bgd.predict(X)

from sklearn.metrics import mean_squared_error
print("mean squared error",mean_squared_error(y, y_pred))

print("weights")
print(bgd.get_weights())
```
