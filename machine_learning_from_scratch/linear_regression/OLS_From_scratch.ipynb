```python
# steps to implement OLS from scratch
# 1) consider a linear equation , the coefficents of the terms will be the true weights which we would predict and check
# 2) make an array X representing the training data with some noise , this array should have the same no. cols as number of variables in the equation considered
# 3) add a column containing all the ones to the original X as the bias
# 4) implement the OLS formula to get the predicted weights
```

```python
# making synthetic data

# considering the true model as : y = 4x1 - 3x2 + 5 + noise , generating data accordingly and then finding the values of weights
import numpy as np
from numpy import random

X = np.random.randn(100,2)
true_weights = np.array([[4],[-3]])  # shape is 2,1
true_bias = 5
noise = np.random.randn(100,1)*0.5

# ground truth (y) corresponding to the considered weights
y = X @ true_weights + true_bias + noise
```

```python
# predicting using the OLS Formula

# add a bias column to the X

ones = np.ones((100,1))
X_new = np.hstack((ones,X))

# Apply OLS formula
# B = (X^t . X )^-1 .X^t Y
predicted_weights = np.linalg.inv(X_new.T @ X_new)@X_new.T@ y
print(predicted_weights)
```
